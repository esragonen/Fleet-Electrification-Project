```python
def yakit_maliyeti_hesapla(row):

    motor_turu = row["Motor türü"]
    yakit_miktari = row["12 ay total yakıt"]

    if motor_turu == "BENZİNLİ":
        return yakit_miktari * 49.58
    elif motor_turu == "DİZEL":
        return yakit_miktari * 51.07
    elif motor_turu == "HYBRID":
        return yakit_miktari * 49.58
    else:
        return 0

df["Yakıt_Maliyeti"] = df.apply(yakit_maliyeti_hesapla, axis=1)

df.head()
```

HYPOTHESIS


H0 : elektrikli ve fosil yakıtlı araçların yıllık yakıt harcamaları arasında anlamlı fark yok

H1 : anlamlı fark var

```python
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.stats.api as sms
from scipy.stats import ttest_1samp, shapiro, levene, ttest_ind, mannwhitneyu, \
    pearsonr, spearmanr, kendalltau, f_oneway, kruskal
from statsmodels.stats.proportion import proportions_ztest
```

```python
test_stat, pvalue = shapiro(df_fiat["yıllık harcama elekt"])
int(pvalue)
```

```python
test_stat, pvalue = shapiro(df_fiat["Yakıt_Maliyeti"])
int(pvalue)
```

Normality yok

```python
test_stat, pvalue = levene(df_fiat["yıllık harcama elekt"], df_fiat["Yakıt_Maliyeti"])
int(pvalue)
```

Variance sağlanmıyor

```python
test_stat, pvalue = mannwhitneyu(df_fiat["yıllık harcama elekt"], df_fiat["Yakıt_Maliyeti"])
int(pvalue)
```

Reject H0. There is a significant difference. Yıllık yakıt harcamaları arasında anlamlı fark var.

```python
final_df = []

for i in range(len(df_citroen)):
    citroen_kar_zarar = df_citroen.loc[i, 'kar/zarar yüzde']
    fiat_kar_zarar = df_fiat.loc[i, 'kar/zarar yüzde']

    if citroen_kar_zarar > fiat_kar_zarar:
        final_df.append(df_citroen.loc[i])
    elif fiat_kar_zarar > citroen_kar_zarar:
        final_df.append(df_fiat.loc[i])
    else:  # Kar/zarar oranları eşitse
        citroen_fiyat = df_citroen.loc[i, 'elektrikli fiyat']
        fiat_fiyat = df_fiat.loc[i, 'elektrikli fiyat']

        if citroen_fiyat <= fiat_fiyat:
            final_df.append(df_citroen.loc[i])
        else:
            final_df.append(df_fiat.loc[i])

final_df = pd.DataFrame(final_df).reset_index(drop=True)

print(final_df.head())
```
